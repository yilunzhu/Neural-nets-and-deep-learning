## Question 1.1
The model excludes those rows that the salary is *N/A*. Following the question to remove all non-numeric features, 16 predictors are kept.
In this assignment, I assume scores has relation to errors. As the score increases, the error rate decreases.

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso, Ridge
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
import math

df = pd.read_csv("Hitters.csv")

non_row = []
non_row += [i for i in range(len(df["Salary"])) if math.isnan(df["Salary"][i])]

df = df.drop(non_row)

y = np.array(df["Salary"], dtype='f')
x_columns = [c for c in df.columns if not isinstance(df[0:1][c].values[0], str)]
x_columns.remove("Salary")
X = np.array(df[x_columns])

folder = 5
kf = KFold(n_splits=folder)

n_alphas = 200
```

```python
alphas = np.logspace(0, 5, n_alphas)
coef = []
scores = []

for a in alphas:
    current_coef = []
    current_scores = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model = Lasso(alpha=a)
        model.fit(X_train, y_train)

        predicted = model.predict(X_test)
        score = model.score(X_test, y_test)

        current_coef.append(model.coef_)
        # print(model.coef_)
        current_scores.append(score)

    current_coef = np.array(current_coef)
    means_coef = [sum(current_coef[:,x])/folder for x in range(len(x_columns))]
    # print(means_coef)
    coef.append(means_coef)
    mean_scores = sum(current_scores) / folder
    scores.append(mean_scores)
```

```python
coef = np.array(coef)
graph = plt.gca()
for n in range(len(x_columns)):
    graph.plot(alphas, coef[:,n], label=x_columns[n])
plt.legend(loc='best')
graph.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficients')
plt.title('Lasso coefficients as a function of the regularization')
plt.axis('tight')
plt.show()

optimal_score = scores.index(max(scores))
left_predicators = 0
for p in coef[optimal_score]:
    if p != 0:
        left_predicators += 1
print("The optimal value of the regularization penalty is 10**%.2f. The number of predictors are left in the model is %s." %(math.log(alphas[optimal_score]), left_predicators))
```

### The bias-variance tradeoff
Bias is how far the predictions are away from actual values. When bias is high, the model is too simple to capture the complexity of the input data. Varivance is to evaluate to what extent the model is fitted into the training data but not the test data. When variance is high, the model peroforms good on the training set but not that good on the test set. This leads to the overfitting problem of the model. With the increasing of alpha, the variance decreases, and the bias increases.

The bias-variance trade-off is finding the balance between bias and variance to minimize errors.

### What role does regularization play in this tradeoff?
In a lienar model, indepedent variables have different weights to predict the output values. In this particular task, the linear regression model combines different weights of several numeric variables to predict the salary. Regularizations penalize the weights of those independent variables. In this task, the model shrinks each weight of those features (such as *"Walks, Years"*). L1 and L2 regularizations try to find a balance between bias and variance, which means a midpoint with low bias and high variance. In short, regularizations do not expect to fit every features to avoid overfitting. However, very few features may cause the model losing the ability to generalize the input data. Therefore, we want regularization to cut off some features to accurately build up the relation between input data and output values.

### High/low bias and variance
- High Bias Low Variance: Models are consistent on both training and test set, but inaccurately predict values.
- High Bias High Variance : Models are inconsistent on both training and test set and also inaccurately predict values.
- Low Bias Low Variance: Models are consistent on both training and test set and accurately predict value.
- Low Bias High variance: Models are inconsistent on both training and test set, but accurately predict values somehow.
